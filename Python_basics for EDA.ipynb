```python
class Person:
    # Class attribute
    species = "Human"
    
    # Constructor
    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age
    
    # Method
    def introduce(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old."
```

```python
# Creating an object
john = Person("John", 30)
print(john.introduce())
```

```python
# Inheritance
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
    
    def study(self):
        return f"{self.name} is studying"
```

```python
# Polymorphism Example
def introduce_person(person):
    print(person.introduce())
```

```python
# Prepare data for visualization
viz_df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [23, 45, 56, 78]
})
```

```python
# 1. Line Chart
plt.figure(figsize=(10, 6))
plt.plot(viz_df['Category'], viz_df['Values'], marker='o')
plt.title('Line Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.savefig('line_chart.png')
plt.show()
```

```python
# 2. Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(viz_df['Category'], viz_df['Values'])
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.savefig('bar_chart.png')
plt.show()
```

```python
# 3. Pie Chart
plt.figure(figsize=(10, 6))
plt.pie(viz_df['Values'], labels=viz_df['Category'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.savefig('pie_chart.png')
plt.show()
```

```python
# 4. Histogram
plt.figure(figsize=(10, 6))
random_data = np.random.normal(0, 1, 1000)
plt.hist(random_data, bins=30)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.savefig('histogram.png')
plt.show()
```

```python
# Seaborn Visualizations
# Prepare Seaborn dataset
tips_df = sns.load_dataset('tips')
```

```python
# 1. Count Plot
plt.figure(figsize=(10, 6))
sns.countplot(x='day', data=tips_df)
plt.title('Count Plot')
plt.savefig('count_plot.png')
plt.show()
```

```python
# 2. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips_df)
plt.title('Box Plot')
plt.savefig('box_plot.png')
plt.show()
```

```python
# 3. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='total_bill', data=tips_df)
plt.title('Violin Plot')
plt.savefig('violin_plot.png')
plt.show()
```

```python
# 4. Pair Plot
plt.figure(figsize=(12, 8))
sns.pairplot(tips_df)
plt.suptitle('Pair Plot', y=1.02)
plt.savefig('pair_plot.png')
plt.show()
```

```python
# 6. Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=tips_df['total_bill'], fill=True)
plt.title('Density Plot')
plt.savefig('density_plot.png')
plt.show()
```
